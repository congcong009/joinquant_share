# 因子处理的相关笔记

## 因子的分析思路
- 关于单因子的结构性分析
    - 单因子分析时最好先按市值三等分，分位高市值、 中市值、小市值。然后再用十分法，把因子值排序，每 10% 取一档。以此可以合理检查分子的单调性，检测其在高市值、中市值、小市值内的稳定性和单调性。
    - 很多因子，在高中小三档市值内的区分度有差别，能在高市值跑得好，不一定能在小市值跑得好；有的因子对市值不敏感（PEG），单调性各有差异，也不一定很稳定。笼统地混为一谈，结论容易有偏差。
    - 最好还能结合行业选股，每个行业千差万别，一刀切，很容易有瑕疵。
- 多重共线性怎么办？
    - 排除引起共线性的变量：找出引起多重共线性的解释变量，将它排除出去，以逐步回归法得到最广泛的应用。
    - 差分法：时间序列数据、线性模型：将原模型变换为差分模型。
    - 减小参数估计量的方差：岭回归法；等等等等。
- 因子池的选择
    - 沪深300可能起作用的因子都是偏价值的，成长系的白马黑马股统统没考虑，妖股小市值也没有入池
## 数据处理
### 因子分层
- 分层的目标是把具有相似基本面的股票收集到一起，不同情景的分层应该能够刻画股票基本面上不同的属性。
- 不同的因子在不同的分层后效果会有显著差异，因此要从不同情景类型中选出对市场风格切换有较强捕捉能力且本身相对稳定的因子，具体来说
    1. 因子的IC正负显著比例之和（percent |t| >2）较高。
    2. 因子横截面间的线性自相关系数较高，自身稳定。
    3. 不同分层因子之间的因子值的横截面秩相关系数较低，能够在不同维度较为独立地刻画股票特征。


### 缺失项
- 财务类：空缺数值 = 本期该字段平均值 * 上期该字段数值 / 上期该字段平均值


## 基本面因子笔记
### 估值和资本结构

#### 账面市值比
- 理解：
    - BP = 1 / pb_ratio
    - pb_ratio 市净率=每股市值（价格）/每股净资产 = price/book value per share，所以 BP = book_value/price
    - pb_ratio 计算，市净率=（股票在指定交易日期的收盘价 * 截至当日公司总股本）/(归属母公司股东的权益MRQ-其他权益工具)，其中，每股净资产＝（资产－负债）/ 公司已发行总股数
    - 如果 BP>1，说明股份对应的公司权益的价值高于股价，也就是说股票的价值被低估了（折价），应该买入。即 BP 越大越好，直观反映出股票估值是否符合公司价值
- 使用：
    - 从小到大排序，买入高分位，卖出低分位
    - 公司包含了制度管理、企业文化、公司商誉、品牌价值等非物质不好度量部分，所以一般会导致公司的市值高于公司净资产价格，即市场公司的平均 BP<1。实证，上海A股为0.59
    - 个股的 BP 指标需要同时结合整个股市的 BP 指标进行解读，BP 受行业影响较大

#### 盈利收益率
- 理解：
    - EP = 1 / pe_ratio
    - pe_ratio 动态市盈率=每股市价 / 每股收益，所以 EP = 每股收益 / 每股价格，可以理解为每股收益在每股交易（机会）成本中的占比
    - 如果公司的 EP 大于市场的平均 EP，可以理解为盈利能力高于平均水平，且价格较低，应该买入。即 EP 越大越好，直观反应出股票被低估的力度。
- 使用：
    - 从小到大排序，买入高分位，卖出低分位
    - EP 与收益回报不是线性关系，因为不代表低估的就能带来超额收益，而只能代表单位资金可能回报会更高
    - 对于价值类的估值偏离是可能被市场修复的，这是趋势，会随着分红等利好操作对股价的提升，产生较好收益，因此高 EP 的买入类似抄底
    - 对于成长类的估值由于短期单位回报不高，但市场对其未来挣钱的预期有认可，即使目前 EP 已经很低了，还会被价格波动拉到更低的水平
    - 针对以上，EP 的使用最好区分成长和价值等风格属性，并按照行业板块进行区分

#### PEG
- 理解：
    - PEG = pe_ratior
    - pe_ratior 市盈率是筛选掉负的市盈率和收益增长率的公司后，与公司的成长性呈负相关，即 PEG 越小，未来的成长性越好。
- 使用：
    - 从小到大排序，买入低分位，卖出高分位

#### 股息率
- 理解：
    - DP = dividend_payable / (market_cap⋅100000000)
    - 股息率 = 公司应付股利 / 总市值
- 使用：
    - 从小到大排序，买入低分位，卖出高分位
    - 与 EP 对比：EP表示公司客观而言盈利的能力（或对于盈利而言估值的水平）。DP 是实际的投资收益比例。
    - 权益收益的定义偏向财务逻辑：如果某投资者年初持有该股票并持有一年，理论上权益收益应该等于该股票的年内价格变动加上股票总派息。但是，在一般财务指标中适用的是，总派息除以当时市价作为投资收益的简化指标。
    - DP 与未来收益的潜在逻辑在于，公司收入水平高则可发股息的空间较大，DP 则低；但高 DP 不如高 EP 一样反映公司更多的盈利能力。
    - 同 EP，也要对于成长性企业，由于高预期带来的高估值影响导致DP过低，尤其是具有高预期的公司可能处于成长初期，自身盈利水平有限，也会在另一方面降低DP指标。因此，低DP策略反映的是高估值选股策略
    - EP 和 DP 同时偏低的股票可能反映的是高估值股票，高估值可能是增长预期推动，具有较好的收益率。

#### 现金收益率
- 理解：
    - CFP = 1 / pcf_ratio
    - pcf_ratio 市现率是每股市价除以过去12个月每股现金流量的倍数，那么现金收益率就是公司现金及等价物与股票市值的比例
    - 据体数据处理过程中使用的是聚宽量化平台的“市现率”指标的倒数，市现率指。
- 使用：
    - 从小到大排序，买入高分位，卖出低分位
    - CFP 主要反映收入结构问题，即收入中现金和等价物的占比
    - 现金等价物虽然是硬指标，但较低的现金流入也可能是由于公司进行了一些长期投资，在一般逻辑中久期的不确定性和非流动性问题会带来更高的平均回报，所以一期的较低的现金流可能并不能说明公司就不好。

#### 市销率
- 理解：
    - PS = ps_ratio
    - 市销率 = 股票价格 / 每股销售收入
    - 市销率越小，通常被认为投资价值越高
- 使用：
    - 从小到大排序，买入低分位，卖出高分位
    - PS 可以在一定程度上反应成长状况，但缺陷在于，销售不包含成本问题，可能存在大量的贴带来的销售；此外也有行业差异性存在，不同行业比较难以得出有意义的结论

#### 资产负债率
- 理解：
    - ARL = total_liability / total_assets
    - 资产负债率 = 总负债 / 总资产
    - ARL 反映了总资产中有多大比例是通过借债来筹资的，也可以在极端情况下分析破产清算时股东和债权人各自的损失情况
    - ARL > 1 就是资不抵债
- 使用：
    - 从小到大排序，买入低分位，卖出高分位
    - 在没有爆发破产前，这个因子作为资本结构因子，主要是衡量公司风险的因子
    - 注意细分债务的构成情况。比如长期负债较高则公司可能再债务方面不需要进行更多频繁处理，短期负债较多则不利企业保持稳定的经济环境。
    - 注重负债比例的变动，一方面扩大生产增大再投资过程中债务比例可能会上升，另一方面自由资产减值也会造成债务比例上升。对于不同情况应该有不同的其他因子组合进入策略，以使得选股策略能够真正进行有效识别。

#### 固定资产比例
- 理解：
    - FACR = fixed_assets / total_assets
    - 固定资产占比 = 公司固定资产 / 总资产。固定资产是指企业为生产商品、提供劳务、出租或经营管理而持有的，使用寿命超过一个会计年度的有形资产，总资产是指企业拥有或可控制的能以货币计量的经济资源
- 使用：
    - 从小到大排序，买入低分位，卖出高分位
    - 传统行业固定资产越低回报率越高，因为较大的 FACR 不利于资金流转盈利。但固定资本占比大也是当前资金使用充分的一个表现，同时较大的固定资本在贷款申请等方面也会产生帮助，因此可能对于抵御波动行情有些许帮助
    - 应该区分行业异质性

#### 流通市值
- 理解：
    - CMC = circulating_market_cap
    - 流通市值（亿元）= 某特定时间内，当时可交易的流通股股数 * 当时股价
- 使用：
    - 从小到大排序，买入低分位，卖出高分位
    - 一般而言
        - 较大的流通股市值，较小的资金量对其操作带来的反身性就很小，交易行为和股票价格不会互相产生太大的影响。较小的反身性带来了正负两方面影响：因为不易炒作，所以很难出现暴涨机会，因此收益不会太高；反之，当股票被大量空单冲击时，造成暴跌的机会更少，便于控制损失
        - 较小流通市值的股票就体现了对立的属性。对于量化投资而言，反身性是比较讨厌的东西，因为会造成较大的冲击成本，导致实际交易和回测结果有很大偏差。
    - 小市值的策略研究参看Fama-French三因子和Fama-French五因子模型。在这两个模型中，小市值被解释为：小规模的公司面临更大的风险，因此它的股票理应有更高的风险补偿。

### 成长因子
#### 净资产收益率
- 理解：
    - ROE = roe
    - ROE (Rate of Return on Common Stockholders’Equity) 净资产收益率，也被称为净值报酬率/权益报酬率/股东权益报酬率/净资产利润率或股权利润率，该指标反映了股东权益的收益水平。
    - ROE = 公司净利润 / 公司股东净资产，指标值越高说明投资带来的收益越高，体现了自有资本获得净收益的能力。这里的净利润指的是“归属于母公司股东的净利润”，剔除了利润中归属于子公司的部分。简化为加权平均净资产收益率，即报告期净利润 / 当期平均资产（期初与期末净资产平均值）
- 使用：
    - 从小到大排序，买入高分位，卖出低分位
    - ROE 缺点：
        - 在没有结合债务指标的情况下，它无法反映出股票定价的合理性。如果一个公司有额外贷款 100 万，那么无贷款的同净资产公司的盈利水平更高并且风险更低，股票定价应该更高，但 ROE 无法反映这个情况
        - 当期如果公司支出过多， ROE 可能表现出较大比例负值。所以 ROE 因子常与其他因子联合使用
        - 不能结合市值变动反映公司当期股价，低 ROE 股票很可能由于 ROE 因子数据不理想导致股价被低估，这也是低 ROE 指标可以有较好收益率的一个合理可能
        - 当期的净收益影响该季度 ROE 指标， ROE 较低可能由于当季度支出（投资性）增加导致，这样的投资性行为可能被视为股价利好刺激。
    - ROE 指标变动较慢，每月调整策略和指标的季报周期有一定差异

#### ROE变动
- 理解：
    - ROEC = ROE上一期 − ROE当期
    - 一般而言公司不出现重大变故，公司股东权益一般变动不会太大，那么 ROE 变动主要来源于当期的净收入水平变动，包括当季度收入的和支出两个方面的影响
- 使用：
    - 从小到大排序，买入高分位，卖出低分位
    - ROEC 只能反映当期的变化，不能对趋势的方向和变化做判断，因此效果不单调
    - ROEC 的问题在于不能识别“不好变到好”还是“好变得更好”，但两种情况对于股票价格冲击是同方向的

#### 资产报酬率
- 理解：
    - $ROA = \frac{net\_profit + income\_tax\_expense + interest\_payable}{total\_sheet\_owner\_equities}$
    - 资产报酬率 = 公司的“净利润”、“所得税费用”和“应付利息” / “负债和股东权益合计”
    - ROA 是衡量公司是否能够有效利用总资金的指标，在实际的财务指标理解中
    - ROE VS ROA 差异性。
        - 在实际应用中，ROE 中分子使用的是“归属于母公司股东的净利润”，而 ROA 使用的收入减去负担的一些非借贷资本成本。
        - ROE 分母使用的是期初和期末平均的归属于母公司的净资产，ROA 分母使用的是期初和期末平均的“负债和股东权益合计”。
        - 分子的差异中 ROA 把借贷资本的成本计入回报之中，而在分母中，ROA 同样把负债纳入指标计算，这样的结果就是公司的财务杠杆的成本和收益统一纳入衡量体系。
        - 无所谓资金是否是借贷而来，所有的自由资金和负债资金的整体运营盈利能力就是 ROA。
        - 两者主要的区别在于对于股东权益和整体公司的侧重，以及财务杠杆作用，因为 ROA 在计算公司盈利能力时同时，除了股东利益外同时把债权人考虑了。假设公司有 10% 的 ROE 并且没有负债，当它增加 1 倍财务杠杆时，便可以使用 2 倍于本身资本的资金进行经济活动，得到的收入扩大一倍，那么在盈利能力不变的情况下 ROE 变为 20%。但是 ROA 考虑了借入资金，所以在增加财务杠杆后 ROA 仍然等于 10%。由此可见 ROA 可以更合理地反应公司的盈利能力。
- 使用：
    - 从小到大排序，买入高分位，卖出低分位
    - ROA 在不同行业体现出较大差异，即需要结合其他原因综合考虑才可以，进一步的处理方法可以利用中性化等处理行业因素影响的方法

#### ROA变动
- 理解：
    - ROAC = ROA上一期 − ROA当期ROAC
    - ROA 变动计算一期为 90 个交易日，与 ROE 变动相类似，ROA 变动也表示企业盈利能力的变动，当然也存在增长和变动两种指标的问题，所以这里使用的时 ROA 变动指标。
    - 与 ROEC 相似，ROAC 指标越大说明企业的运营能力变得越差，而 ROAC 越小表示变得越好。
- 使用：
    - 从小到大排序，买入低分位，卖出高分位
    - 同 ROEC，多因子和该因子的改良处理，这是后续优化结果的两个选择。

#### 主营业务利润占比
- 理解：
    - $OPTP = \frac{operating\_profit}{total\_profit}$
    - 主营业务利润占比 = 营业利润 / 利润总额
    - 主营利润占比较高，说明企业主要工作中心在主营业务上，两方面影响：
        - 从好的一面说，主营业务占比较高的企业工作重心集中，在有限的企业资源条件约束下集中，企业应该主要依托主营业务发展
        - 不好的一面而言，主营业务依赖度较高，对于行业层面的系统性风险抵抗能力较弱，需要依赖行业良好的发展条件企业自身才有较好的发展空间
- 使用：
    - 从小到大排序，买入高分位，卖出低分位
    - 作为财务指标，公司的主营业务策略对于公司的影响常常超过季度财报的周期，应尽可能在季度进行仓位调整，因为 OPTP 不能表达较短时间的市场反映，股价也不能良好的捕获 OPTP 指标带来的中长期市值增幅回报

#### 主营利润占收入比例
-理解：
    - $OPOR = \frac{operating_profit}{operating_revenue}$
    - OPOR 与 OPTP 差异:
        - OPOR 使用的分母是“营业收入”，而 OPTP 分母使用的使“总利润”。
        - 对比财务细节，两者的差距主要在于总利润与营业收入的范围差异。利润总额指企业在生产经营过程中各种收入扣除各种耗费后的盈余，反映企业在报告期内实现的盈亏总额。但是营业收入中包含了没计算在利润中的很多成本和支出。
        - 营业收入包括主营业务收入和其他业务收入，一般而言其他业业务收入较低。
        - 营业收入与利润中间的差值包含也营业成本和业务税金及附加，OPOR 较小意味着差值部分较大，那么这样的公司可能本身利润水平不算高，通过高速的经营周转速度带来较高最终利润水平。
- 使用：
    - 从小到大排序，买入低分位，卖出高分位
    - OPOR 反向选取的很可能出现类似零售业等高周转低销售利润率（单位销售物品的利润率）的产业，应注意区分

#### 净利率占收入比例
- 理解：
    - $NPOR = \frac{net_profit}{operating_revenue}$
    - 收入净利率 = 净利润 / 主营收入
    - 与 OPOR 相比主要差别在于，分子选用了净利润而不是主营利润，相比 OPOR 分子使用的营业利润，净利润指标表示的更加纯粹，剔除了营业外费用的影响。
- 使用：
    - 从小到大排序，买入高分位，卖出低分位
    - 可以看出增长因子中很多对于财务数据指标的选取都大同小异，主要在于不同的指标可能有有不同的适用场景。

### 技术因子
#### 月动能
- 理解：
    - $MTM1 = \frac{close当期 − close一月前}{close一月前}$
    - 单纯的动量策略使用往期回报数据来预测未来截面回报数据，所谓未来截面回报（ Cross Section of Future Return） 讲的是一个时间点上不同投资产品的预测回报率。
    - 动量策略就是在上一个观测的窗口期计算回报率，然后计算排序，通过排序选择回报率最高的几只股票并持有。
    - “动量”和“反转”是一对绕不开的冤家，如果上述策略中选择的是回报排名最低几只就变成反转策略了，反转的直观理解就是跌的太多了要反弹。
- 使用：
    - 设计思路：
        - 观测的窗口期定多长时间
        - 选择多少只股票
        - 动量还是反转，决定了到底选择最高的几只还是回报最低的几只
        - 调仓周期选择
        - 其他，如是否动态调仓、每只股票仓位选择、调仓时入场时间和价格判断和出场条件等
    - 长期动量变现稍好于短期，因为逻辑上理解，长期股票回报率较好的公司，应该会由较大几率带来更高回报

#### 换手率
- 理解：
    - $TR = \frac{\sum{volume}}{circulating\_cap⋅10000}$
    - 换手率 = 一个时间窗口内（一个月）的股票交易数量 / 期末的的流通股票数量。
    - 换手率是一定时间内市场中股票转手买卖的频率，是反映股票流通性强弱的指标之一。
    - 如果一支股票的换手率较高，说明这支股票表现的比较活跃，投资者对它的参与（关注）度的较高。
    - 从交易的逻辑上而言，供过于求则价格上涨，供小于求则价格下降。那么一只股票的换手率过低表明交易双方参与较少，可以理解为这样的一个价位上，价格较为被市场认可，因此没有过多的交易产生。

#### 波动率
- 理解：
    - $Volatility = VAR(\frac{close_t}{close_t−1}-1)$
    - （月）波动率 = 过去的一个月中股票的日收益率的方差
    - 波动率一般用来测度股票价格的风险，实际上两日的收益率可以看作价格变动的一阶关系描述，那么波动率就是价格变动的二阶描述。
    - 二阶条件描述的是价格变动的变化情况，价格变动的累计加总（累计正负两个方向的价格变动）描述了股票价格的总体不确定性，而总体不确定性被用术语“风险”表述。
    - 指标越大表示在过去的约一个月中，该股票的价格波动越大，越小表明该股票上月价格波动越小。
- 使用：
    - CAPM 理论解释证券的很多风险都是可以靠合理配置中和掉，那么既然能被中和掉就不应该由于风险补偿而产生收益，而真正产生风险补偿的是不能被中和的部分。基于此，高波动率的股票不仅承受了太多“多余”的风险，也没有获得风险补偿，所以收益自然就较低了。

#### 震荡
- 理解：
    - $FLUC = \frac{close_{max}−close_{min}}{close_{期初}+close_{期末}}$
    - 根据市场经验：横有多长，竖有多高。就直觉而言，简单表述为月波动范围 / 月平均价格乘以2，是方差与动量两者的结合产物
    - 该因子一定大于 0，那么该因子越大表明本月震荡幅度越大，越小表明本月震荡越小
    - 但这个因子表现一般

#### 翻转因子
- 理解：
    - $Rate_{i,t-\eta } = \frac{P^{close}_{i,t-\eta}}{P^{10:30}_{i,t-\eta}} $
    - $Reversers_{i,t}=\prod_{\eta = 0}^{21} Rate_{i,t-\eta}$
    - Rate：股票当日10：30到收盘收益率，P：股票价格，close：收盘时间，即15：00，i：第i支股票，t：当前时间，$\eta$：回溯时间，目前业界普遍使用的是22个交易日（一个自然月）
